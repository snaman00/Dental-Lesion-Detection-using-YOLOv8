```python
%%writefile "/kaggle/working/main.py"
import os
import logging
import torch
import gc
from datetime import datetime
from ultralytics import YOLO
import yaml

def setup_logging():
    log_dir = "logs"
    os.makedirs(log_dir, exist_ok=True)
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    log_file = os.path.join(log_dir, f"training_{timestamp}.log")

    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s',
        handlers=[
            logging.FileHandler(log_file),
            logging.StreamHandler()
        ]
    )
    return logging.getLogger(__name__)

def check_gpu_memory():
    """Check GPU memory status"""
    if torch.cuda.is_available():
        gpu_memory = torch.cuda.get_device_properties(0).total_memory
        memory_allocated = torch.cuda.memory_allocated(0)
        return f"Total: {gpu_memory / 1e9:.2f}GB, Used: {memory_allocated / 1e9:.2f}GB"
    return None

def main():
    logger = setup_logging()
    logger.info("Starting dental cavity detection training pipeline")

    if not torch.cuda.is_available():
        logger.error("CUDA is not available. Please ensure you have GPU runtime enabled in Colab")
        return

    gpu_info = check_gpu_memory()
    logger.info(f"GPU Memory Status: {gpu_info}")

    try:
        # Path to Training Configuration and data.yaml file
        config_path = '/kaggle/working/configs/training_config.yaml'
        data_path = '/kaggle/working/data.yaml'

        with open(config_path, 'r') as f:
            config = yaml.safe_load(f)

        # Initialize model you want to train,
        model = YOLO('yolov8l.pt')
        logger.info("Model loaded successfully")

        augmentations = {
            'hsv_h': 0.015,
            'hsv_s': 0.7,
            'hsv_v': 0.4,
            'degrees': 10.0,
            'translate': 0.1,
            'scale': 0.5,
            'shear': 2.0,
            'perspective': 0.0005,
            'flipud': 0.0,
            'fliplr': 0.5,
            'mosaic': 1.0,
            'mixup': 0.2
        }

        logger.info("Starting Stage 1 training with YOLOv8l")
        results_stage1 = model.train(
            data=data_path,
            epochs=config['training']['epochs']['stage1'],
            imgsz=config['dataset']['img_size'],
            batch=config['dataset']['batch_size'],
            optimizer=config['training']['optimizer'],
            lr0=config['training']['learning_rate'],
            weight_decay=config['training']['weight_decay'],
            warmup_epochs=config['training']['warmup_epochs'],
            save=True,
            project='runs',
            name='stage1_l',
            device=0, 
            exist_ok=True,
            save_period=10,
            **augmentations
        )
        logger.info(f"Stage 1 completed: {results_stage1}")

    except Exception as e:
        logger.error(f"Training failed: {str(e)}")
        logger.exception("Detailed error traceback:")

if __name__ == "__main__":
    main()
```
